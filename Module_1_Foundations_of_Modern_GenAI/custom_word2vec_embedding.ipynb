```python
folder_path = 'data'

story = []

for file in os.listdir(folder_path):
    file_path = os.path.join(folder_path, file)
    with open(file=file_path, mode='r', encoding='utf-8') as f:
        corpus = f.read()

    sentences = sent_tokenize(corpus)
    for sentence in sentences:
        story.append(simple_preprocess(sentence))
        
```

## Custom Model

```python
custom_model = gensim.models.Word2Vec(
    window=10,
    min_count=1,
    vector_size=150
    sg=0            #sg=0 for CBOW, sg=1 for Skip-gram
)
```

```python
custom_model.build_vocab(story)
```

```python
len(custom_model.wv.index_to_key)
```

```python
custom_model.corpus_count
```

```python
custom_model.train(story, total_examples=custom_model.corpus_count, epochs=10)
```
