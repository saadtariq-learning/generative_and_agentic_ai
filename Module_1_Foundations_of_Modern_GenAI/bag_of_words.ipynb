```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
documents = [
    "people watch movie and watch movie again",
    "people watch football and watch football",
    "people like movie and like movie a lot",
    "people like football",
]
```

```python
bag_of_words = CountVectorizer()
```

```python
bag_of_words_matrix = bag_of_words.fit_transform(documents)
bag_of_words_matrix.toarray()
```

```python
print("Vocabulary:", bag_of_words.get_feature_names_out())
```
