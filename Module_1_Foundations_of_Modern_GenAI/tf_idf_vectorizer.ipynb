```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
documents = [
    "people watch football",
    "football watch football",
    "people give comment",
    "football give comment",
]

second_documents = [
    "people watch movie and watch movie again",
    "people watch football and watch football",
    "people like movie and like movie a lot",
    "people like football",
]
```

```python
documents
```

```python
tf_idf = TfidfVectorizer()
```

## For 1st Document

```python
tf_idf_vectorizer = tf_idf.fit_transform(documents)
```

```python
for vector in tf_idf_vectorizer.toarray():
    print(vector)
```

```python
tf_idf.get_feature_names_out()
```

```python
print("Vocabulary:", tf_idf.get_feature_names_out())
```

## For 2nd Document

```python
tf_idf_vectorizer = tf_idf.fit_transform(second_documents)
```

```python
for vector in tf_idf_vectorizer.toarray():
    print(vector)
```

```python
tf_idf.get_feature_names_out()
```

```python
print("Vocabulary:", tf_idf.get_feature_names_out())
```
