## Accessing Gemini LLMs

```python
import os
import base64
import wave
from google import genai
from google.genai import types
from langchain_google_genai import ChatGoogleGenerativeAI
from langchain_core.messages import HumanMessage, SystemMessage
```

```python
GOOGLE_API_KEY = os.environ.get("GOOGLE_API_KEY")

if GOOGLE_API_KEY:
    os.environ["GOOGLE_API_KEY"] = GOOGLE_API_KEY
```

### Text to Text

```python
model = ChatGoogleGenerativeAI(
    model="gemini-3.5-flash"
)
```

```python
messages = [
    SystemMessage(content="You are a helpful assistant that translates English to French. Translate the user sentence."),
    HumanMessage(content="I love programming.")
]
```

```python
response = model.invoke(messages)
print(response.content)
```

```python
response.content[0]['text']
```

### Image to Text Model

```python
def encode_image(image_path):
    with open(image_path, "rb") as image_file:
        return base64.b64encode(image_file.read()).decode("utf-8")

base64_image = encode_image("images/funny_dog.jpeg")
```

```python
llm = ChatGoogleGenerativeAI(
    model="gemini-3.5-flash",
)
```

```python
message = HumanMessage(
    content=[
        {
            "type": "text",
            "text": "Describe this image in detail."
        },
        {
            "type": "image_url",
            "image_url": f"data:image/jpeg;base64,{base64_image}"
        }
    ]
)
```

```python
response = llm.invoke([message])
print(response.content[0]['text'])
```

### Audio to Text

```python
def encode_audio(audio_path):
    with open(audio_path, "rb") as audio_file:
        return base64.b64encode(audio_file.read()).decode("utf-8")
```

```python
base64_audio = encode_audio("audios/audio.mp3")
```

```python
llm = ChatGoogleGenerativeAI(
    model="gemini-3.5-flash",
)
```

```python
message = HumanMessage(
    content=[
        {
            "type": "text",
            "text": "Transcribe this audio into English text."
        },
        {
            "type": "media",
            "mime_type": "audio/mp3",
            "data": base64_audio
        }
    ]
)
```

```python
response = llm.invoke([message])
print(response.content[0]['text'])
```

### Text to Audio

```python
client = genai.Client(api_key=GOOGLE_API_KEY)
```

```python
response = client.models.generate_content(
    model="gemini-2.5-flash-preview-tts", #freely accessible model for text to speech conversion
    contents="Hello students, today we are learning multimodal AI, which is a very interesting topic.",
    config=types.GenerateContentConfig(
        response_modalities=["AUDIO"],
        speech_config=types.SpeechConfig(
            voice_config=types.VoiceConfig(
                prebuilt_voice_config=types.PrebuiltVoiceConfig(
                    voice_name="Kore"
                )
            )
        )
    )
)
```

```python
audio_data = response.candidates[0].content.parts[0].inline_data.data
```

```python
with wave.open("audios/output.wav", "wb") as wf:
    wf.setnchannels(1)
    wf.setsampwidth(2)
    wf.setframerate(24000)
    wf.writeframes(audio_data)

print("Audio saved as output.wav")
```
